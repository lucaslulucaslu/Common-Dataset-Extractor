```python
import pandas as pd
```

```python
df=pd.read_csv("GPAs_part_1.csv")
df2=pd.read_csv("GPAs.csv")
```

```python
df=pd.concat([df,df2])
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df_same=df[df['Name']==df['ExtractName']]
```

```python
df_same=df_same.drop(df_same[df_same['GPA400']=='0'].index)
```

```python
df_same=df_same.drop(df_same[(df_same['GPA400']=="0%") & (df_same['GPA375']=="0%")& (df_same['GPA350']=="0%")& (df_same['GPA325']=="0%")& (df_same['GPA300']=="0%")& (df_same['GPA250']=="0%")& (df_same['GPA200']=="0%")& (df_same['GPA100']=="0%")& (df_same['GPA000']=="0%")].index)
```

```python
df_same=df_same.drop(df_same[(df_same['GPA400']=="0.00%") & (df_same['GPA375']=="0.00%")& (df_same['GPA350']=="0.00%")& (df_same['GPA325']=="0.00%")& (df_same['GPA300']=="0.00%")& (df_same['GPA250']=="0.00%")& (df_same['GPA200']=="0.00%")& (df_same['GPA100']=="0.00%")& (df_same['GPA000']=="0.00%")].index)
```

```python
df_same=df_same.drop(df_same[df_same['GPA400'].isnull()&df_same['GPA375'].isnull()&df_same['GPA350'].isnull()&df_same['GPA325'].isnull()&df_same['GPA300'].isnull()&df_same['GPA250'].isnull()&df_same['GPA200'].isnull()&df_same['GPA100'].isnull()&df_same['GPA000'].isnull()].index)
```

```python
df_same
```

```python
df_same=df_same.drop(df_same[df_same['GPA400'].apply(lambda x:len(str(x))>10)].index)
```

```python
df_same.reset_index(drop=True,inplace=True)
```

```python
df_same.to_csv("after.csv",index=False)
```
